# Requirements
1.	Thiết lập mô hình toán học tổng quát của các xung:
+ Xung vuông (square wave)
+ Xung tam giác (triagle wave)
+ Xung răng cưa (sawtooth wave)
+ Xung chuỗi (pules train)
Bằng mô hình chuỗi Fourier với tần số f = 1000*sin(2*pi/49*N) N: Số thứ tự của sinh viên
2.	Viết code để thực hiện vẽ các xung đã thiết lập mô hình toán học:
+ khảo sát với 1000 sóng?
+ khảo sát với góc pha phi=2*pi/N? 
+ Tìm biên độ của xung?
+ Vẽ thống kê biên độ của hàm tổng hợp (các xung) theo tổng số sóng thành phần chạy từ 100 đến 1000?


```python
import numpy as np
import matplotlib.pyplot as plt

# Tính toán tần số f dựa trên công thức đã cho
N = 27
f = 1000 * np.sin(2 * np.pi / 49 * N)

# Góc pha
phi = 2 * np.pi / N
```

# Square wave
1. **Xung vuông (Square wave)**:
   $$  f(t) = \frac{4}{\pi} \sum_{k=1,3,5,...}^{\infty} \frac{\sin(2\pi kft)}{k}  = \frac{4}{\pi} \sum_{k=1,3,5,...}^{\infty} \frac{\sin(1.2\pi kt)}{k} $$

```python
# Hàm tạo xung vuông dựa trên chuỗi Fourier
def square_wave(t, N, phi):
    return 4/np.pi * sum(np.sin(2 * np.pi * (2*k-1) * f * t + phi) / (2*k-1) for k in range(1, N+1))

# Khởi tạo mảng thời gian
t = np.linspace(-0.01, 0.01, 1000)  # Điều chỉnh khoảng thời gian để phù hợp với xung

# Vẽ xung vuông
plt.figure(figsize=(10, 4))
plt.plot(t, square_wave(t, 1000, phi))
plt.title('Xung vuông (Square Wave)')
plt.xlabel('Thời gian (s)')
plt.ylabel('Biên độ')
plt.grid(True)
plt.show()
```

```python

# Tìm biên độ của xung
amplitude = np.max(square_wave(t, 1000, phi))
print(f'Biên độ của xung vuông: {amplitude}')

# Vẽ thống kê biên độ theo tổng số sóng thành phần
amplitudes = [np.max(square_wave(t, n, phi)) for n in range(100, 1001)]
plt.figure(figsize=(10, 4))
plt.plot(range(100, 1001), amplitudes)
plt.title('Thống kê biên độ của xung vuông theo tổng số sóng thành phần')
plt.xlabel('Số sóng thành phần')
plt.ylabel('Biên độ')
plt.grid(True)
plt.show()
```

# Triangle wave
2. **Xung tam giác (Triangle wave)**:
   $$  f(t) = \frac{8}{\pi^2} \sum_{k=1,3,5,...}^{\infty} \frac{(-1)^{(k-1)/2}\sin(1.2\pi kt)}{k^2}  $$

```python
# Hàm tạo xung tam giác dựa trên chuỗi Fourier
def triangle_wave(t, N, phi):
    return (8 / np.pi**2) * sum((-1)**((k-1)//2) * np.sin(2 * np.pi * k * f * t + phi) / k**2 for k in range(1, N+1, 2))

# Khởi tạo mảng thời gian
t = np.linspace(-0.01, 0.01, 1000)  # Điều chỉnh khoảng thời gian để phù hợp với xung

# Vẽ xung tam giác
plt.figure(figsize=(10, 4))
plt.plot(t, triangle_wave(t, 1000, phi))
plt.title('Xung tam giác (Triangle Wave)')
plt.xlabel('Thời gian (s)')
plt.ylabel('Biên độ')
plt.grid(True)
plt.show()

# Tìm biên độ của xung
amplitude = np.max(triangle_wave(t, 1000, phi))
print(f'Biên độ của xung tam giác: {amplitude}')

# Vẽ thống kê biên độ theo tổng số sóng thành phần
amplitudes = [np.max(triangle_wave(t, n, phi)) for n in range(100, 1001)]
plt.figure(figsize=(10, 4))
plt.plot(range(100, 1001), amplitudes)
plt.title('Thống kê biên độ của xung tam giác theo tổng số sóng thành phần')
plt.xlabel('Số sóng thành phần')
plt.ylabel('Biên độ')
plt.grid(True)
plt.show()
```

# Sawtooth wave
3. **Xung răng cưa (Sawtooth wave)**:
   $$  f(t) = \frac{2}{\pi} \sum_{k=1}^{\infty} \frac{(-1)^{k+1}\sin(2\pi kft)}{k} = \frac{2}{\pi} \sum_{k=1}^{\infty} \frac{(-1)^{k+1}\sin(1.2\pi kt)}{k} $$

```python
# Hàm tạo xung răng cưa dựa trên chuỗi Fourier
def sawtooth_wave(t, N, phi):
    return (2 / np.pi) * sum((-1)**(k+1) * np.sin(2 * np.pi * k * f * t + phi) / k for k in range(1, N+1))

# Khởi tạo mảng thời gian
t = np.linspace(-0.01, 0.01, 1000)  # Điều chỉnh khoảng thời gian để phù hợp với xung

# Vẽ xung răng cưa
plt.figure(figsize=(10, 4))
plt.plot(t, sawtooth_wave(t, 1000, phi))
plt.title('Xung răng cưa (Sawtooth Wave)')
plt.xlabel('Thời gian (s)')
plt.ylabel('Biên độ')
plt.grid(True)
plt.show()

# Tìm biên độ của xung
amplitude = np.max(sawtooth_wave(t, 1000, phi))
print(f'Biên độ của xung răng cưa: {amplitude}')

# Vẽ thống kê biên độ theo tổng số sóng thành phần
amplitudes = [np.max(sawtooth_wave(t, n, phi)) for n in range(100, 1001)]
plt.figure(figsize=(10, 4))
plt.plot(range(100, 1001), amplitudes)
plt.title('Thống kê biên độ của xung răng cưa theo tổng số sóng thành phần')
plt.xlabel('Số sóng thành phần')
plt.ylabel('Biên độ')
plt.grid(True)
plt.show()
```

# Pulse train
4. **Xung chuỗi (Pulse train)**:
$$
\text{pulse\_train}(t, N, \phi, \tau, T) = \sum_{n=-N}^{N} \sin(2 \pi n f t + \phi) \cdot \text{sinc}(n \tau f)
$$

Params:
- $t$: Time variable.
- $N$: Number of harmonics.
- $\phi$: Phase shift.
- $\tau$: Pulse width.
- $T$: Period of the pulse train.
- $f$: Frequency of the pulse train (not provided in your code but necessary for the formula).

```python
# Hàm tạo xung chuỗi dựa trên chuỗi Fourier
def pulse_train(t, N, phi, tau, T):
    return sum(np.sin(2 * np.pi * n * f * t + phi) * np.sinc(n * tau * f) for n in range(-N, N+1))

# Khởi tạo mảng thời gian
t = np.linspace(-0.01, 0.01, 1000)  # Điều chỉnh khoảng thời gian để phù hợp với xung

# Độ rộng xung khoảng cách giữa các xung
phi = 2 * np.pi / N
tau = 0.1  # Độ rộng xung
T = 0.5  # Khoảng cách giữa các xung

# Vẽ xung chuỗi
plt.figure(figsize=(10, 4))
plt.plot(t, pulse_train(t, 1000, phi, tau, T))
plt.title('Xung chuỗi (Pulse Train)')
plt.xlabel('Thời gian (s)')
plt.ylabel('Biên độ')
plt.grid(True)
plt.show()

# Tìm biên độ của xung
amplitude = np.max(pulse_train(t, 1000, phi, tau, T))
print(f'Biên độ của xung chuỗi: {amplitude}')

# Vẽ thống kê biên độ theo tổng số sóng thành phần
amplitudes = [np.max(pulse_train(t, n, phi, tau, T)) for n in range(100, 1001)]
plt.figure(figsize=(10, 4))
plt.plot(range(100, 1001), amplitudes)
plt.title('Thống kê biên độ của xung chuỗi theo tổng số sóng thành phần')
plt.xlabel('Số sóng thành phần')
plt.ylabel('Biên độ')
plt.grid(True)
plt.show()
```
