```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
## SIN1: Create an array of x values from 0 to 2 pi
t1 = np.arange(0, 2, 0.01)
y1 = np.sin(2 * np.pi * t1) # Calculate the sine of each x value

## SIN2 
t2 = np.arange(0.0, 2.0, 0.01)
y2 = np.sin(2 * np.pi * t2 + np.pi /2)

# Plot the sine wave
plt.plot(t1, y1, t2, y2)

# Add labels and a title
plt.xlabel('time (s)')
plt.ylabel('Y-axis')
plt.title('Sine wave Plot')

# Add Grid
plt.grid()

# Display the plot
plt.show()
```
